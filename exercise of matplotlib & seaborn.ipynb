```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
df = pd.read_excel('C:/Users/panke/Downloads/WFU/HR analysis/book.xlsx',header = 0, index_col = 'id')
df.head(5)
```

```python
df.educ.plot(kind = 'hist',alpha = 0.5, title = 'Education Histogram')
plt.xlabel('Education Year')
plt.show()

df.AgeVariable.plot(kind = 'hist', alpha = 0.5, title = 'Age Histogram')
plt.xlabel('Age Year')
plt.show()

df.plot.scatter(x = 'Weeks', y = 'ScanPminute', c = 'green', alpha = 0.5, title = 'plot of weeks in work and scanminute')
plt.show()


df[df['disability'] == 0].ScanPminute.plot( kind = 'box',color = 'red' )
df[df['disability'] == 1].ScanPminute.plot( kind = 'box',color = 'blue')
plt.legend(['No Disability', 'Have Disability'],loc = 'upper left')
plt.show()

df[df['Training'] == 1].plot.scatter(x = 'AgeVariable', y = 'Improvement', alpha = 0.5, title = 'Age and Improvement',
                                    marker = 's')
plt.xlabel('Age Year')
plt.ylabel('Improvement Number')
plt.show()

```

```python
import seaborn as sns
sns.set(style='ticks')

np.random.seed(0)
N = 37
_genders= ['Female', 'Male', 'Non-binary', 'No Response']
df11 = pd.DataFrame({
    'Height (cm)': np.random.uniform(low=130, high=200, size=N),
    'Weight (kg)': np.random.uniform(low=30, high=100, size=N),
    'Gender': np.random.choice(_genders, size=N)
})

fg = sns.FacetGrid(data=df11, hue='Gender', hue_order=_genders, aspect=1.61)
fg.map(plt.scatter, 'Weight (kg)', 'Height (cm)').add_legend()
```

```python
carat = [5, 10, 20, 30, 5, 10, 20, 30, 5, 10, 20, 30]
price = [100, 100, 200, 200, 300, 300, 400, 400, 500, 500, 600, 600]
color =['D', 'D', 'D', 'E', 'E', 'E', 'F', 'F', 'F', 'G', 'G', 'G',]

df = pd.DataFrame(dict(carat=carat, price=price, color=color))

fig, ax = plt.subplots()

colors = {'D':'red', 'E':'blue', 'F':'green', 'G':'black'}

ax.scatter(df['carat'], df['price'], c=df['color'].apply(lambda x: colors[x]))

plt.show()
```
